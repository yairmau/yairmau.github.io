## Changing Lightness of Matplotlib Color

```python
import matplotlib.colors
import colorsys
import numpy as np
def add_lightness(color_name, L):
    """
    Receives a color name ("red", "tab:blue", "xkcd:green", etc),
    Returns RGB values of same color with lightness L added to it (can be negative too)
    """
    color_hls = colorsys.rgb_to_hls(*matplotlib.colors.to_rgb(color_name))
    new_color_hls = np.array(color_hls)
    new_color_hls[1] = new_color_hls[1] + L
    if new_color_hls[1] > 1.0: new_color_hls[1] = 1.0
    if new_color_hls[1] < 0.0: new_color_hls[1] = 0.0
    new_shade_rgb = colorsys.hls_to_rgb(*new_color_hls)
    return new_shade_rgb

add_lightness("gold", 0.1)
```

## xkcd colors that I like

Don't use them all together. Use black, some shade of gray, and one or two of the colors below. That's more than enough.
Source: [xkcd color survey](https://xkcd.com/color/rgb/){target="_blank"}

````python
#| code-summary: "colors that I like"
#| code-fold: true
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
c_list = ["xkcd:goldenrod", "xkcd:hot pink", "xkcd:forest", "xkcd:cerulean", "xkcd:vermillion", "xkcd:grass green", "xkcd:barney", "xkcd:royal blue"]
n_colors = len(c_list)
np.random.seed(4)
N = 40
fig, ax = plt.subplots(1, 2, figsize=(10, 3))
for i, c in enumerate(c_list):
    d1 = (np.random.rand(N)-0.5).cumsum(); d1=d1-d1[0]
    d2 = np.random.normal(0, 0.5, N).cumsum(); d2=d2-d2[0]
    ax[0].plot(d1[::-1]+2*i, color=c, linewidth=5)
    ax[1].plot(d2+2*i, color=c, linewidth=5)
    ax[0].text(N*1.13, 2*i, c[5:], fontsize=8, color=c, va="center", ha="center")
    hex_code = mcolors.to_hex(mcolors.to_rgb(c))
    print(f"{c} = {hex_code}")
ax[0].set_axis_off()
ax[1].set_axis_off()
ax[0].set(ylim=(-3, 2*n_colors+2))
ax[1].set(ylim=(-3, 2*n_colors+2));
````
